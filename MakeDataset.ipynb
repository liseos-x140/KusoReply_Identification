# 教師データ作成用の対話文を決定

```python
import pickle
from random import choices
```

```python
corpus_path = "corpus/"
filename_input = corpus_path + "input.pkl"
filename_output = corpus_path + "output.pkl"
```

```python
with open(filename_input, "rb")as f:
    uttarances = pickle.load(f)
    
with open(filename_output, "rb")as f:
    responses = pickle.load(f)
```

重複を消す(bot対策) + 発話と返信に分割

```python
_tmp = [_s.split("<pause>") for _s in set(["<pause>".join([_u,_r]) for _u,_r in zip(uttarances,responses)])]
uttarances = [_t[0] for _t in _tmp]
responses = [_t[1] for _t in _tmp]
```

## 1000対の対話文

```python
index = choices(list(range(len(uttarances))), k=1000)

dataset_u = [_u for i,_u in enumerate(uttarances) if i in index]
dataset_r = [_r for i,_r in enumerate(responses) if i in index]
```

### 表示

```python
for i,(_u,_r) in enumerate(zip(dataset_u,dataset_r)):
    print(_u, "\n", _r, "\n")
```

### Save

```python
dataset_input = corpus_path + "dataset_U.pkl"
dataset_output = corpus_path + "dataset_R.pkl"

with open(dataset_input, "wb")as f:
    pickle.dump(dataset_u, f)
    
with open(dataset_output, "wb")as f:
    pickle.dump(dataset_r, f)
```
